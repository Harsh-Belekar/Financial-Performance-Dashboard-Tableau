```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('../data/financial_data.csv')
```

```python
# Clean Column Names
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.columns
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Replace '$-' with '$0'
df['Discounts'] = df['Discounts'].str.replace('$-', '$0', regex=False)
df['Profit'] = df['Profit'].str.replace('$-', '$0', regex=False)
```

```python
# Clean Currency & Numeric Columns
currency_cols = [
    'Units_Sold', 'Manufacturing_Price', 'Sale_Price',
    'Gross_Sales', 'Discounts', 'Sales', 'COGS','Profit'
]

for col in currency_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('(', '', regex=False)
        .str.replace(')', '', regex=False)
        .str.strip()
        .astype(float)
    )
```
